```python
def convert_to_numeric(value, type):
    if type == 'Current_loan_status':
        if value == 'NO DEFAULT':
            return 0
        elif value == 'DEFAULT':
            return 1
        else:
            return None  # Manejo de valores inesperados
    elif type == 'historical_default':
        if value == 'Y':
            return 1
        elif value == 'N':
            return 0
        else:
            return None  # Manejo de valores inesperados
    return value
      
# Aplicar las conversiones a números
df2['Current_loan_status'] = df2['Current_loan_status'].apply(lambda x: convert_to_numeric(x, 'Current_loan_status'))
df2['historical_default'] = df2['historical_default'].apply(lambda x: convert_to_numeric(x, 'historical_default'))
```

```python
df2['customer_income'] = df2['customer_income'].replace('[\£,]', '', regex=True).astype(float)

from sklearn.impute import KNNImputer

# Crear una instancia del KNNImputer
imputer = KNNImputer(n_neighbors=2)

# Imputar los valores nulos
df2[['historical_default',
      'customer_age',
      'customer_income',
       'term_years',
       'cred_hist_length',
       'Current_loan_status']] = imputer.fit_transform(df2[['historical_default',
                                                            'customer_age',
                                                            'customer_income',
                                                            'term_years',
                                                            'cred_hist_length',
                                                            'Current_loan_status']])

 # Asegurarse de que los valores imputados son enteros (0 o 1)
df2['historical_default'] = df2['historical_default'].round().astype(int)
df2['Current_loan_status'] = df2['Current_loan_status'].round().astype(int)
print(df2['Current_loan_status'].value_counts())
print(df2['historical_default'].value_counts())
```

```python
def convert_status(value, type):
    if type == 'Current_loan_status':
        if value == 0:
            return 'NO DEFAULT'
        elif value == 1:
            return 'DEFAULT'
        else:
            return None  # Manejo de valores inesperados
    elif type == 'historical_default':
        if value == 1:
            return 'Y'
        elif value == 0:
            return 'N'
        else:
            return None  # Manejo de valores inesperados
    return value

# Aplicar las conversiones a objetos
df2['Current_loan_status'] = df2['Current_loan_status'].apply(lambda x: convert_status(x, 'Current_loan_status'))
df2['historical_default'] = df2['historical_default'].apply(lambda x: convert_status(x, 'historical_default'))
```
