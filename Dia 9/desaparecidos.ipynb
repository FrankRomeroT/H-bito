```python
df = pd.read_csv('C:/Users/FRANK/Desktop/Proyectos Linkedin/Python Proyectos/Dia 9/MM_14_21.csv')
df.head(3)
```

```python
df2 = df.copy()
print('Número de observaciones (filas) = {}'.format(df2.shape[0]))
print('Total de variables disponibles (columnas) = {} '.format(df2.columns.shape[0]))
print('Columnas que cuentan con un valor nulo -')
print(df2.isnull().sum())
```

```python
# Cambiamos el nombre Unnamed:0 a ID y agregar valores secuenciales ala columna ID
df2.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)
df2['ID'] = range(1,len(df2)+1)
df2 = df2.drop(columns=['Main ID','Incident ID','URL','Incident Date','Coordinates'])
```

```python
df2['Source Quality'].value_counts()
df2['Number Dead'].value_counts()
```

```python
df2[['Number Dead','Source Quality']].corr()
```

```python
# Aqui se uso interpolación porque son numeros secuenciales, es decir continuos, por lo tanto la interpolación aqui es recomendable 

df2['Number Dead']= df2['Number Dead'].interpolate().round()
df2['Source Quality']= df2['Source Quality'].interpolate().round()
```

```python
df2.info()
```

```python
print('Columnas que cuentan con un valor nulo -')
print(df2.isnull().sum())
```

```python
df2['Information Source'] = df2['Information Source'].fillna('Desconocido')
```

```python
df2.isnull().sum()
```

```python
# Se puede observar que hay una correlación alta entre | Minimum Estimated Number of Missing , Number of Males y Number of Children | ,
# por lo tanto se podria recomendar una interpolación conjunta , sin embargo usare K Vecinos Más Cercanos
df2[['Minimum Estimated Number of Missing','Number of Survivors','Number of Females','Number of Males','Number of Children']].corr()
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2)

columna_impute = ['Minimum Estimated Number of Missing','Number of Children','Number of Males']
df2[columna_impute] = imputer.fit_transform(df2[columna_impute]).round()
```

```python
df2['UNSD Geographical Grouping'] = df2['UNSD Geographical Grouping'].fillna('Northern America')
```
