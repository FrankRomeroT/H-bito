```python
x1 = pd.read_csv('C:/Users/FRANK/Desktop/Proyectos Linkedin/Python Proyectos/Dia 7/euro_summary.csv', sep=';')
x2 = pd.read_csv('C:/Users/FRANK/Desktop/Proyectos Linkedin/Python Proyectos/Dia 7/Uefa Euro Cup General Statistics.csv', sep=';')
x3 = pd.read_csv('C:/Users/FRANK/Desktop/Proyectos Linkedin/Python Proyectos/Dia 7/Uefa Euro Cup Participated Teams General Statistics.csv')
x1.rename(columns={'Year': 'Year'}, inplace=True)
# Fusionar x1 y x2 usando la columna 'Year'
merged_1_2 = pd.merge(x1, x2, on='Year', how='outer')
```

```python
merged_1_2 = merged_1_2.drop(columns=['Winning coach','Top scorer(s)','Player of the Tournament','result'])
```

```python
merged_1_2.loc[15,'Champion'] = 'Italy' 
merged_1_2.loc[15,'Host'] = 'England'
merged_1_2.loc[16,'Host'] = 'Germany'
merged_1_2[['winner', 'final']] = merged_1_2[['winner', 'final']].replace('USSR', 'Russia')
```

```python
# Remplazamos la palabra USSR por Rusia y Soviet Union por Rusia
merged_1_2[['team1', 'team2']] = merged_1_2['final'].str.split(' - ', expand=True)
merged_1_2[['team1','team2']] = merged_1_2[['team1','team2']].replace('USSR','Russia')
merged_1_2[['team1','team2']] = merged_1_2[['team1','team2']].replace('USSR','Russia')
merged_1_2[['Champion']] = merged_1_2[['Champion']].replace('Soviet Union','Russia')
# Eliminamos la columna final
merged_1_2 = merged_1_2.drop(columns=['final'])

# Remplazamos la palabra West Germany por Germany
merged_1_2[['winner','Host','Champion','team1','team2']] = merged_1_2[['winner','Host','Champion','team1','team2']].replace('West Germany','Germany')
```
