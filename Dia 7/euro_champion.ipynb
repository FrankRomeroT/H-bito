# Regresión logistica multimonial

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
# Asegurarse de que las columnas existen en merged_1_2
merged_1_2.columns = merged_1_2.columns.str.strip()
enc_column = ['winner', 'team1', 'team2']
target_column = 'Champion'

# Separar los datos en conjuntos de entrenamiento y prueba, considerando solo las filas sin valores nulos en 'Champion'
train_data = merged_1_2[merged_1_2[target_column].notna()]
test_data = merged_1_2[merged_1_2[target_column].isna()]

# Preprocesar los datos
encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
X_train = encoder.fit_transform(train_data[enc_column])
X_test = encoder.transform(test_data[enc_column])
y_train = train_data[target_column]

# Crear y entrenar el modelo de regresión logística multinomial
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
model.fit(X_train, y_train)

# Predecir los valores nulos de 'Champion'
y_pred = model.predict(X_test)

# Llenar los valores nulos en el DataFrame original
merged_1_2.loc[merged_1_2[target_column].isna(), target_column] = y_pred

print(merged_1_2)
```

```python
# Asumimos que tienes una copia del DataFrame original sin la imputación
original_merged_1_2 = merged_1_2.copy()

# Asegurarse de que las columnas existen en merged_1_2
merged_1_2.columns = merged_1_2.columns.str.strip()
enc_column = ['winner', 'team1', 'team2']
target_column = 'Champion'

# Verificar si hay valores nulos en la columna 'Champion'
null_values = original_merged_1_2[original_merged_1_2[target_column].isna()]
if null_values.empty:
    print("No hay valores nulos en la columna 'Champion'.")
else:
    # Separar los datos en conjuntos de entrenamiento y prueba, considerando solo las filas sin valores nulos en 'Champion'
    train_data = original_merged_1_2[original_merged_1_2[target_column].notna()]
    test_data = original_merged_1_2[original_merged_1_2[target_column].isna()]

    # Preprocesar los datos
    encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
    X_train = encoder.fit_transform(train_data[enc_column])
    X_test = encoder.transform(test_data[enc_column])
    y_train = train_data[target_column]

    # Crear y entrenar el modelo de regresión logística multinomial
    model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
    model.fit(X_train, y_train)

    # Predecir los valores nulos de 'Champion'
    y_pred = model.predict(X_test)

    # Llenar los valores nulos en el DataFrame original
    merged_1_2.loc[merged_1_2[target_column].isna(), target_column] = y_pred

    # Identificar las filas imputadas
    imputed_rows = original_merged_1_2[original_merged_1_2[target_column].isna()].copy()
    imputed_rows['Predicted_Champion'] = y_pred

    # Mostrar las filas imputadas con el valor predicho de 'Champion'
    print(imputed_rows[['Year', 'winner', 'team1', 'team2', 'Predicted_Champion']])
```
