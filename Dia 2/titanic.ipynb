```python
df2['Age'] = df2['Age'].round().fillna(0).astype(int)
df2.head()
```

```python
df2['Age'] = df2['Age'].interpolate(method='linear')
df2.head()
```

```python
df2['Embarked'] = df2['Embarked'].fillna(df2['Embarked'].mode()[0])
df2['Embarked'].value_counts()
```

```python
df2 = df2.drop(columns=['Cabin'])
```

# Analisis

```python
# Categorizar las edades en los tres grupos: <18, 18-50, >50
age_categories = pd.cut(df2['Age'], bins=[0, 18, 50, 100], labels=['<18', '18-50', '>50'])

# Contar el número de personas en cada categoría de edad
age_counts = age_categories.value_counts().sort_index()

# Crear el gráfico de pie
plt.figure(figsize=(8, 8))
age_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Distribución de Edades de los Pasajeros del Titanic')
plt.ylabel('')  # Ocultar la etiqueta del eje y

# Mostrar el gráfico
plt.show()
```

```python
gender_counts = df2['Sex'].value_counts()
plt.pie(gender_counts , labels = gender_counts.index, autopct='%1.1f%%', colors=['Blue', 'Red'], startangle=140)
plt.title('Porcentaje de Hombres y Mujeres')
plt.show()
```

```python
df2.groupby('Sex')['Survived'].sum().plot(kind='bar', color='Blue')
plt.title('Titanic')
plt.xlabel('Sobrevivio')
plt.ylabel('')
plt.xticks(rotation=0)
plt.show()
```

```python
df2.groupby(['Pclass', 'Survived'])['Survived'].count()
ax=sns.countplot(x='Pclass', hue='Survived', palette='Set1', data=df2)
ax.set(title='Estado del pasajero (murio/sobrevivio) dado a la clase a la que pertenecia', 
       xlabel='Clase del pasajero | 0 = Murio | 1 = Sobrevivio|', ylabel='Total')
plt.show()
```

```python
ax = sns.catplot(x='Pclass',hue = 'Sex', col = 'Survived',palette = 'Set1', data=df2, kind='count')
plt.show()
```
