```python
import matplotlib.pyplot as plt
import numpy as np

# Nmero de entradas y unidades en las capas ocultas
num_inputs = 21
num_units_hidden = 64

# Crear las posiciones de las entradas y las unidades de la capa oculta
input_positions = np.linspace(0, 1, num_inputs)
hidden_positions = np.linspace(0, 1, num_units_hidden)

fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar las entradas
for i, pos in enumerate(input_positions):
    ax.plot([0, 0.5], [pos, pos], 'k-', lw=1)
    ax.text(-0.1, pos, f'Input {i+1}', horizontalalignment='center', verticalalignment='center')

# Dibujar las unidades de la capa oculta
for i, pos in enumerate(hidden_positions):
    ax.plot([0.5, 1], [pos, pos], 'k-', lw=1)
    ax.text(1.1, pos, f'Hidden {i+1}', horizontalalignment='center', verticalalignment='center')

# Conectar entradas con unidades de la capa oculta
for input_pos in input_positions:
    for hidden_pos in hidden_positions:
        ax.plot([0, 0.5, 1], [input_pos, hidden_pos, hidden_pos], 'k-', lw=0.1)

ax.set_xlim(-0.2, 1.2)
ax.set_ylim(-0.1, 1.1)
ax.axis('off')

plt.title('Visualizacin de la Red Neuronal: 21 Entradas a 64 Unidades en la Capa Oculta')
plt.show()
```
