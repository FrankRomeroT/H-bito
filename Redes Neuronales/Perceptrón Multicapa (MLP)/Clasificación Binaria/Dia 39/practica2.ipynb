```python
from sklearn.preprocessing import MinMaxScaler, PowerTransformer
import numpy as np
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
import random

# Generar datos exponenciales
df2 = [random.expovariate(0.2) for _ in range(1000)]
df2 = np.array(df2).reshape(-1, 1)  # Convertir la lista en un array de NumPy

# Realizar prueba de Shapiro-Wilk en los datos originales
w_orig, p_orig = stats.shapiro(df2)
print("Shapiro-Wilk test para datos originales:", w_orig, p_orig)

# Graficar los datos originales
sns.histplot(df2, kde=True)
plt.title("Datos originales")
plt.show()

# Aplicar MinMaxScaler
scaler = MinMaxScaler()
df2_normalizado = scaler.fit_transform(df2)

# Realizar prueba de Shapiro-Wilk en los datos normalizados
w_norm, p_norm = stats.shapiro(df2_normalizado)
print("Shapiro-Wilk test para datos normalizados:", w_norm, p_norm)

# Graficar los datos normalizados
sns.histplot(df2_normalizado, kde=True)
plt.title("Datos normalizados con MinMaxScaler")
plt.show()

```

```python
from sklearn.preprocessing import PowerTransformer

# Aplicar transformacin Yeo-Johnson
pt = PowerTransformer(method='yeo-johnson')
df2_transformado = pt.fit_transform(df2)

# Realizar prueba de Shapiro-Wilk en los datos transformados
w_trans, p_trans = stats.shapiro(df2_transformado)
print("Shapiro-Wilk test para datos transformados:", w_trans, p_trans)

# Graficar los datos transformados
sns.histplot(df2_transformado, kde=True)
plt.title("Datos transformados con Yeo-Johnson")
plt.show()
```
