```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos ficticios para los ejemplos
epochs = np.arange(1, 51)

# Datos para cada escenario
excelente_train_loss = np.exp(-epochs / 10) + 0.01 * np.random.randn(len(epochs))
excelente_val_loss = np.exp(-epochs / 10) + 0.01 * np.random.randn(len(epochs))

bien_train_loss = np.exp(-epochs / 10) + 0.05 * np.random.randn(len(epochs))
bien_val_loss = np.exp(-epochs / 15) + 0.05 * np.random.randn(len(epochs))

regular_train_loss = np.exp(-epochs / 15) + 0.1 * np.random.randn(len(epochs))
regular_val_loss = np.exp(-epochs / 20) + 0.1 * np.random.randn(len(epochs))

malo_train_loss = np.exp(-epochs / 20) + 0.1 * np.random.randn(len(epochs))
malo_val_loss = np.exp(-epochs / 25) + 0.15 * np.random.randn(len(epochs))

muy_malo_train_loss = np.exp(-epochs / 30) + 0.15 * np.random.randn(len(epochs))
muy_malo_val_loss = np.exp(-epochs / 35) + 0.2 * np.random.randn(len(epochs))

# Crear una figura con subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 18))

# Excelente
axes[0, 0].plot(epochs, excelente_train_loss, label='Entrenamiento')
axes[0, 0].plot(epochs, excelente_val_loss, label='Validación')
axes[0, 0].set_title('Excelente')
axes[0, 0].set_xlabel('Épocas')
axes[0, 0].set_ylabel('Pérdida')
axes[0, 0].legend()

# Bien
axes[0, 1].plot(epochs, bien_train_loss, label='Entrenamiento')
axes[0, 1].plot(epochs, bien_val_loss, label='Validación')
axes[0, 1].set_title('Bien')
axes[0, 1].set_xlabel('Épocas')
axes[0, 1].set_ylabel('Pérdida')
axes[0, 1].legend()

# Regular
axes[1, 0].plot(epochs, regular_train_loss, label='Entrenamiento')
axes[1, 0].plot(epochs, regular_val_loss, label='Validación')
axes[1, 0].set_title('Regular')
axes[1, 0].set_xlabel('Épocas')
axes[1, 0].set_ylabel('Pérdida')
axes[1, 0].legend()

# Malo
axes[1, 1].plot(epochs, malo_train_loss, label='Entrenamiento')
axes[1, 1].plot(epochs, malo_val_loss, label='Validación')
axes[1, 1].set_title('Malo')
axes[1, 1].set_xlabel('Épocas')
axes[1, 1].set_ylabel('Pérdida')
axes[1, 1].legend()

# Muy Malo
axes[2, 0].plot(epochs, muy_malo_train_loss, label='Entrenamiento')
axes[2, 0].plot(epochs, muy_malo_val_loss, label='Validación')
axes[2, 0].set_title('Muy Malo')
axes[2, 0].set_xlabel('Épocas')
axes[2, 0].set_ylabel('Pérdida')
axes[2, 0].legend()

# Ajustar el diseño y ocultar el subplot vacío
plt.tight_layout()
axes[2, 1].axis('off')
plt.show()
```
