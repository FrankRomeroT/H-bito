```python
import numpy as np
from scipy import stats
import statsmodels.api as sm

# Generar datos de ejemplo
data = np.random.normal(0, 1, 1000)

# Prueba de Shapiro-Wilk
shapiro_test = stats.shapiro(data)
print(f'Shapiro-Wilk Test: Estadístico = {shapiro_test.statistic}, Valor p = {shapiro_test.pvalue}')

# Prueba de Kolmogorov-Smirnov
ks_test = stats.kstest(data, 'norm')
print(f'Kolmogorov-Smirnov Test: Estadístico = {ks_test.statistic}, Valor p = {ks_test.pvalue}')

# Prueba de Anderson-Darling
ad_test = stats.anderson(data, dist='norm')
print(f'Anderson-Darling Test: Estadístico = {ad_test.statistic}')
print(f'Valores críticos = {ad_test.critical_values}')
print(f'Niveles de significancia = {ad_test.significance_level}')

# Prueba de Lilliefors
lilliefors_test = sm.stats.lilliefors(data)
print(f'Lilliefors Test: Estadístico = {lilliefors_test[0]}, Valor p = {lilliefors_test[1]}')

# Prueba de Jarque-Bera
jb_test = stats.jarque_bera(data)
print(f'Jarque-Bera Test: Estadístico = {jb_test.statistic}, Valor p = {jb_test.pvalue}')

# Prueba de D'Agostino y Pearson
dp_test = stats.normaltest(data)
print(f'D\'Agostino-Pearson Test: Estadístico = {dp_test.statistic}, Valor p = {dp_test.pvalue}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Generar datos de ejemplo
data = np.random.normal(0, 1, 1000)

# Histograma y gráfica de densidad
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True)
plt.title('Histograma y Gráfica de Densidad')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import statsmodels.api as sm

# Gráfica QQ
plt.figure(figsize=(10, 6))
sm.qqplot(data, line='45', fit=True)
plt.title('Gráfica QQ')
plt.show()
```

```python
# Realizar los tests de normalidad
shapiro_test = stats.shapiro(data)
ks_test = stats.kstest(data, 'norm')
ad_test = stats.anderson(data, dist='norm')
lilliefors_test = sm.stats.lilliefors(data)
jb_test = stats.jarque_bera(data)
dp_test = stats.normaltest(data)

# Recolectar los p-valores
test_names = ['Shapiro-Wilk', 'Kolmogorov-Smirnov', 'Anderson-Darling', 'Lilliefors', 'Jarque-Bera', 'D\'Agostino-Pearson']
p_values = [shapiro_test.pvalue, ks_test.pvalue, lilliefors_test[1], jb_test.pvalue, dp_test.pvalue]

# Gráfica de P-Valores
plt.figure(figsize=(10, 6))
sns.barplot(x=test_names, y=p_values)
plt.axhline(y=0.05, color='r', linestyle='--')
plt.title('P-Valores de Tests de Normalidad')
plt.ylabel('P-Valor')
plt.xlabel('Test de Normalidad')
plt.show()
```
